```python
df_gastos_varios = pd.read_excel('03_MARZO_2022.xlsx', sheet_name='GTOS VARIOS', skiprows=5)
df_gastos_varios['SHEET'] = 'GTOS VARIOS'
```

```python
df_servicios = pd.read_excel('03_MARZO_2022.xlsx', sheet_name='SERV_PROF', skiprows=5)
df_servicios['SHEET'] = 'SERV_PROF'
```

```python
df_comunicaciones = pd.read_excel('03_MARZO_2022.xlsx', sheet_name='COMUNICACION', skiprows=5)
df_comunicaciones['SHEET'] = 'COMUNICACION'
```

```python
df_comunicaciones.head()
```

```python
df_personales = pd.read_excel('03_MARZO_2022.xlsx', sheet_name='SERV PERS', skiprows=5)
df_personales['SHEET'] = 'SERV PERS'
```

```python
df_personales.head()
```

```python
df_compras_marzo = pd.concat([df_gastos_varios, df_contratistas, df_servicios, df_comunicaciones, df_personales])
```

```python
df_compras_marzo.head()
```

```python
df_compras_marzo.shape
```

```python
#Borramos las filas con valores faltantes
df_compras_marzo_com=df_compras_marzo.dropna(how='any')
```

```python
df_compras_marzo_com.shape
```

```python
df_compras_marzo_com.columns
```

```python
#verificamos si los datos faltante aun existen
list(df_compras_marzo_com['CVE BEN'])
```

```python
# verificamos el tipo de datos que tenemos en cada columna
df_compras_marzo_com.dtypes
```

```python
df_compras_marzo_com['FECHA\nPAGO'].head()
```

```python
#transformar la columna FECHA\nPAGO a datetime64
df_compras_marzo_com['FECHA\nPAGO'] = df_compras_marzo_com['FECHA\nPAGO'].apply(lambda x: pd.to_datetime(x, format='%Y/%m/%d'))
```

```python
#verificamos el tipo de datos que tenemos
df_compras_marzo_com.dtypes
```

```python
#verificamos que las fechas sean de un solo mes
df_compras_marzo_com['FECHA\nPAGO'].apply(lambda x: x.month).value_counts()
```

```python
#eliminamos las fechas con valor month != 3
df_compras_marzo_com = df_compras_marzo_com[df_compras_marzo_com['FECHA\nPAGO'].apply(lambda x: x.month) == 3]
```

```python
#verificamos nuevamente que esten en un solo mes las fechas
df_compras_marzo_com['FECHA\nPAGO'].apply(lambda x: x.month).value_counts()
```
